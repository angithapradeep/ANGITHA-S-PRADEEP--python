```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arra = np.arange(1,10).reshape(3,3)
for i in range(len(arra)):
    plt.plot(arra[i])
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)


arra1 = np.arange(1, 26)
ax1.plot(arra1,arra1**2, color = 'green')
arra2 = np.arange(25, 1, -1)
ax2.plot(arra2[::-1],arra2**2, color ='red')
ax3.plot((arra1**2)*-1,arra1, color = 'blue')
ax4.plot(arra2[::-1],(arra2**2)*-1, color = 'orange')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
x1 = np.arange(-10,10,0.1)
y1 = np.sin(x1)
x2 = np.arange(-10, 10, 0.1)
y2 = np.cos(x2)
x3=x1
y3 = np.tan(x3)
ax1.plot(x3,y3, "b")
ax3.plot(x1, y1, "r-")
ax4.plot(x2,y2, "g-")
plt.show()
```

```python
t = np.linspace(0, 2 * np.pi, 1000)

x = 16 * np.sin(t)**3 # x = 16*sin^3(t)
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t) # y = 13*cos(t) - 5*cos(2t) - 2*cos(3t) - cos(4t)

plt.plot(x, y, color='red')
plt.title('❤️ Heart Curve')
plt.axis('equal')
plt.show()
```
